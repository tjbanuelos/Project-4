```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
MLB = pd.read_csv(
    "Resources/MLB.csv")

# Display sample data
MLB.head(10)
```

```python
# Generate summary statistics
MLB.describe()
```

```python
# Plot your data to see what's in your DataFrame
MLB.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file


MLB_scaled = StandardScaler().fit_transform(MLB[[
    'player_id', 'year', 'player_age',  # changed from 'player_aged'
    'pa', 'k_percent',                  # changed from 'pk_percent'
    'bb_percent', 'batting_avg', 'slg_percent', 'on_base_percent',
    'on_base_plus_slg', 'woba', 'exit_velocity_avg',
    'barrel_batted_rate', 'hard_hit_percent', 'avg_best_speed',
    'iz_contact_percent', 'swing_percent', 'pull_percent',
    'straightaway_percent', 'opposite_percent', 'sprint_speed'
]])
```

```python
# Create a DataFrame with the scaled data 
df_mlb_scaled = pd.DataFrame(MLB_scaled, 
                             columns=['player_id','year', 'player_age', 'pa', 'k_percent', 'bb_percent',
                                      'batting_avg', 'slg_percent', 'on_base_percent', 'on_base_plus_slg',
                                      'woba', 'exit_velocity_avg', 'barrel_batted_rate', 'hard_hit_percent',
                                      'avg_best_speed', 'iz_contact_percent', 'swing_percent', 'pull_percent',
                                      'straightaway_percent', 'opposite_percent', 'sprint_speed'])

# Add player_id from the index
df_mlb_scaled['player_id'] = MLB.index

# Set 'player_id' as index
df_mlb_scaled = df_mlb_scaled.set_index('player_id')

# Display
df_mlb_scaled.head(10)

```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k=list(range(1,11))
```

```python
df_mlb_scaled_clean = df_mlb_scaled.dropna()
```

```python
# Create an empty list to store the inertia values
inertia=[]

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data 
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(df_mlb_scaled_clean)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k_original": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_original = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_plot = df_elbow_original.hvplot.line(
    x="k_original", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)

elbow_plot
```

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=2, random_state=0)
```

```python
# Fit the K-Means model using the scaled DataFrame
model.fit(df_mlb_scaled_clean)
```

```python
# Predict the clusters to group the MLB using the scaled DataFrame
kmeans_predictions = model.predict(df_mlb_scaled_clean)


# Print the resulting array of cluster values.
kmeans_predictions
```

```python
# Create a copy of the scaled DataFrame
MLB_scaled_prediction=df_mlb_scaled_clean.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
MLB_scaled_prediction['Prediction']=kmeans_predictions

# Display the copy of the scaled DataFrame
MLB_scaled_prediction.head()
```

```python
# Apply PCA to reduce to 2 dimensions for visualization
pca = PCA(n_components=2)
MLB_pca = pca.fit_transform(df_mlb_scaled_clean)

# Create a new DataFrame with PCA results
MLB_pca_df = pd.DataFrame(MLB_pca, columns=["PC1", "PC2"])
MLB_pca_df["last_name, first_name"] = MLB["last_name, first_name"]
MLB_pca_df["player_age"] = MLB["player_age"]
MLB_pca_df["batting_avg"] = MLB["batting_avg"]
```

```python
# Run KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
MLB_pca_df["Prediction"] = kmeans.fit_predict(df_mlb_scaled_clean)
```

```python
# Visualize PCA components with clusters
MLB_pca_df.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="Prediction",
    hover_cols=["last_name, first_name", "player_age", "batting_avg"],
    width=700,
    height=500,
    title="MLB Player Clustering by Performance Metrics"
)
```

```python
# Add predictions to original DataFrame
MLB["Prediction"] = MLB_pca_df["Prediction"]

# Visualize clusters by Exit Velocity and Sprint Speed
MLB.hvplot.scatter(
    x="exit_velocity_avg",
    y="sprint_speed",
    by="Prediction",
    hover_cols=["last_name, first_name", "player_age", "batting_avg"],
    width=700,
    height=500,
    title="MLB Clusters: Exit Velocity vs Sprint Speed"
)
```
