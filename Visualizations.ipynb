```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("data/raw_data/statcast_batter_stats_2023_2025.csv")

# Drop missing values for relevant columns
bat_contact_df = df.dropna(subset=['avg_best_speed', 'iz_contact_percent'])
# 4. Define x and y
x = bat_contact_df['avg_best_speed']
y = bat_contact_df['iz_contact_percent']

# 5. Fit linear regression (trendline)
slope, intercept = np.polyfit(x, y, 1)
trendline = slope * x + intercept

# 6. Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.6, edgecolor='k', label='Players')
plt.plot(x, trendline, color='red', linewidth=2, label='Trendline')

# 7. Labels and styling
plt.title('Bat Speed vs. In-Zone Contact Percentage')
plt.xlabel('Average Best Bat Speed (mph)')
plt.ylabel('In-Zone Contact %')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


#Load the data
df = pd.read_csv("data/raw_data/statcast_batter_stats_2023_2025.csv") 

#Compute correlation matrix for numeric columns only
corr_matrix = df.corr(numeric_only=True)

#Plot heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(
    corr_matrix,
    cmap="coolwarm",
    annot=False,
    linewidths=0.5,
    square=True,
    cbar_kws={"shrink": 0.8}
)
plt.title("Correlation Heatmap of Statcast Batting Metrics")
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("data/raw_data/statcast_batter_stats_2023_2025.csv")  
#  Drop rows with missing values for exit velocity and barrel rate
ev_barrel_df = df.dropna(subset=['exit_velocity_avg', 'barrel_batted_rate'])

# Prepare variables
x = ev_barrel_df['exit_velocity_avg']
y = ev_barrel_df['barrel_batted_rate']

#  Calculate and plot trendline
slope, intercept = np.polyfit(x, y, 1)
trendline = slope * x + intercept

#  Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.6, edgecolor='k', label='Players')
plt.plot(x, trendline, color='red', linewidth=2, label='Trendline')

#  Add titles and labels
plt.title('Exit Velocity vs. Barrel Rate')
plt.xlabel('Average Exit Velocity (mph)')
plt.ylabel('Barrel Rate (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("data/raw_data/statcast_batter_stats_2023_2025.csv")  # adjust path as needed

#Group by year and calculate average OPS
ops_by_year = df.groupby('year')['on_base_plus_slg'].mean().reset_index()
ops_by_year['year'] = ops_by_year['year'].astype(int)

#Plot bar chart
plt.figure(figsize=(8, 5))
plt.bar(ops_by_year['year'], ops_by_year['on_base_plus_slg'], color='salmon', edgecolor='black')
plt.title('Average OPS by Year')
plt.xlabel('Year')
plt.ylabel('Average OPS')
plt.ylim(0.5,0.75)  # OPS typically ranges from ~0.5 to ~1.2
plt.xticks(ops_by_year['year'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("data/raw_data/statcast_batter_stats_2023_2025.csv")  # adjust path as needed

# Drop rows with missing OPS values
df = df.dropna(subset=['on_base_plus_slg'])

# Define OPS bins and labels
bins = [0, 0.599, 0.699, 0.799, 0.899, float('inf')]
labels = ['< .600', '.600.699', '.700.799', '.800.899', '.900+']

# Bin the OPS data
df['OPS_range'] = pd.cut(df['on_base_plus_slg'], bins=bins, labels=labels)

# Count how many players fall into each OPS range
ops_counts = df['OPS_range'].value_counts().sort_index()

# Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(ops_counts, labels=ops_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Player OPS (20232025)')
plt.axis('equal')  # Ensures the pie is circular
plt.tight_layout()
plt.show()
```
